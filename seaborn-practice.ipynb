# Seaborn 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_theme(style="darkgrid")
```

## 1. Loading Seaborn Sample Datasets

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips'); df.head()
```

## 2. Relational Plots

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(data=df, x="size", y="total_bill")
plt.show()
```

```python
sns.relplot(data=df, x='total_bill', y='tip', col='sex')
```

## 3. Categorical Plots

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='day', y='total_bill')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='day')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='day', y='tip')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(data=df, x='day', y='total_bill')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.stripplot(data=df, x='day', y='tip')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.swarmplot(data=df, x='day', y='tip')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.pointplot(data=df, x='day', y='tip')
plt.show()
```

```python
sns.catplot(data=df, x='day', y='tip', kind='box', col='sex')
```

## 4. Distribution Plots

```python
sns.histplot(df['total_bill'], kde=True)
```

```python
sns.kdeplot(df['total_bill'])
```

```python
sns.ecdfplot(df['total_bill'])
```

```python
sns.displot(df['total_bill'], kde=True)
```

```python
sns.jointplot(data=df, x='total_bill', y='tip', kind='scatter')
```

```python
sns.pairplot(df[['total_bill','tip','size']])
```

## 5. Regression Plots

```python
plt.figure(figsize=(6,4))
sns.regplot(data=df, x="total_bill", y="tip")
plt.show()
```

```python
sns.lmplot(data=df, x='total_bill', y='tip', col='sex')
```

## 6. Matrix Plots

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
sns.clustermap(df.corr(), annot=True)
```

## 7. Styling in Seaborn

```python
sns.set_theme(style='whitegrid')
sns.set_palette('husl')
sns.set_context('talk')
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='day', y='total_bill')
plt.show()
```

## 8. FacetGrid Examples

```python
g = sns.FacetGrid(df, col="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
```

## 9. Combining Seaborn + Matplotlib

```python
ax = sns.barplot(data=df, x="day", y="tip")
ax.set_title("Tips per Day")
ax.set_ylabel("Tip Amount")
plt.show()
```
